# BoardGameGeek - Items

El fichero `bgg_items.csv` contiene los datos básicos de cada uno de los juegos que se han extraido de la web [BoardGameGeek](https://boardgamegeek.com/). Los atributos incluidos en este fichero son los siguientes:

- `id`: Identifficador del juego.
- `name`: Nombre del juego.
- `year`: Año de publicación del juego.
- `is-expansion`: 1 si el registro es una expansion, 0 si es un juego base.
- `minplayers`: Mínimo número de jugadores indicados en el juego.
- `maxplayers`: Máximo número de jugadores indicados en el juego.
- `bestminnumplayers`: Mejor número mínimo de jugadores.
- `bestmaxnumplayers`: Mejor número máximo de jugadores.
- `rminnumplayers`: Mínimo número de jugadores recomendados por la comunidad.
- `rmaxnumplayers`: Máximo número de jugadores recomendados por la comunidad.
- `age`: Edad mínima recomendada para el juego.
- `averageweight`: Representa la complejidad del juego siendo 1 el más fácil y 5 el más difícil.
- `minplaytime`: Duración mínima de una partida.
- `maxplaytime`: Duración máxmima de una partida.
- `url`: URL de la fiche del juego en BGG.

```python
# Definimos las columnas que queremos que se guarden en el csv
bgg_items_cols = ['id', 'name',  'year', 'expansion', 'minplayers', 'maxplayers', 'bestminnumplayers',
       'bestmaxnumplayers', 'rminnumplayers', 'rmaxnumplayers', 'age', 'averageweight', 'minplaytime', 'maxplaytime',  
       'url']
```

```python
bgg_items_df = df[bgg_items_cols]
bgg_items_df.to_csv('../data/bgg_items.csv', index=False)
```
